### Method 3

```python
import numpy as np

print(np.arange(1, 7))    # Takes default steps of 1 and doesn't include 7
print(np.arange(5))       # Starts at 0 by default and ends at 4, giving 5 numbers
print(np.arange(1, 10, 3))  # Starts at 1 and ends at less than 10, with a step size of 3
```

### Method 4

```python
import numpy as np

print(np.linspace(1, 12, 12))
print(np.linspace(1, 12, 5))
print(np.linspace(1, 12, 3))
```

# Basic Plotting
Source: [Python for Scientists and Engineers](https://www.educative.io/courses/python-for-scientists-and-engineers/JYJwDK58Kp9)

## Example 1

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 100)
y = x ** 2
fig = plt.figure(figsize=(10, 5), dpi=100)
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])     #add_axes([bottom x, bottom y, width, height])
ax1.plot(x, y, 'b')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Example figure');
```

## Example 2

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
fig = plt.figure(figsize = (10, 5))
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax1.plot(x, y, color='g', marker='o', linestyle='-', linewidth=4, markersize=15)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('y = sin(x)');
```
